```python
%%time


for x in Country:
    for y in Product:
        for z in Label_Code:
            
            

#####################################################################################################             
# Print current loop and time
#####################################################################################################             
            time2 = datetime.datetime.now()
            time2=time2.strftime("%Y")+time2.strftime("%m")+time2.strftime("%d")+"_"+time2.strftime("%H")+time2.strftime("%M")
            print(time2)
            print(x)
            print(y)
            print(z)
#####################################################################################################
# Prepare Input
#####################################################################################################  


            # Create dataframes one for with and one for without labels. Loop over country and product;
            if x == "Total":
                df = df_all.query(f"Product==@y and Label=='{label_map[z]}'")          

            else:
            # Create dataframes one for with and one for without labels. Loop over country and product; 
                df = df_all.query(f"Country==@x and Product==@y and Label=='{label_map[z]}'")
            
             # Only keep columns relevant for Biogeme Estimation
            if z == "WoL":
                df =df[['RID_New', 'BEST', 'WORST', 'a1_x1', 'a1_x2', 'A1_PRICE', 'a1_x3', 'A1_MAT_CC', 'A1_MAT_OC', 'A1_MAT_WO', 'A1_MAT_RM', 'A1_MAT_PE', 'a1_x4', 'A1_COM_CHN', 'A1_COM_EUR', 'A1_COM_BGD', 'A1_COM_VNM', 'A1_COM_A_G', 'a1_x5', 'A1_Q_LOW', 'A1_Q_MED', 'A1_Q_HIGH', 'a2_x1', 'a2_x2', 'A2_PRICE', 'a2_x3', 'A2_MAT_CC', 'A2_MAT_OC', 'A2_MAT_WO', 'A2_MAT_RM', 'A2_MAT_PE', 'a2_x4', 'A2_COM_CHN', 'A2_COM_EUR', 'A2_COM_BGD', 'A2_COM_VNM', 'A2_COM_A_G', 'a2_x5', 'A2_Q_LOW', 'A2_Q_MED', 'A2_Q_HIGH', 'a3_x1', 'a3_x2', 'a3_x3', 'a3_x4', 'a3_x5']]
           
            else:
                df = df[['RID_New', 'BEST', 'WORST', 'a1_x1', 'a1_x2', 'A1_PRICE', 'a1_x3', 'A1_MAT_CC', 'A1_MAT_OC', 'A1_MAT_WO', 'A1_MAT_RM', 'A1_MAT_PE', 'a1_x4', 'A1_COM_CHN', 'A1_COM_EUR', 'A1_COM_BGD', 'A1_COM_VNM', 'A1_COM_A_G', 'a1_x5', 'A1_Q_LOW', 'A1_Q_MED', 'A1_Q_HIGH', 'a1_x6', 'A1_Label', 'A1_L1', 'A1_L1_E_CO2', 'A1_L1_E_WAT', 'A1_L1_E_WAS', 'A1_L1_E_ENE', 'A1_L1_S_WFS', 'A1_L1_S_HRS', 'A1_L1_G_MGS', 'A1_L1_G_CSR', 'A1_L2', 'A1_L2_E', 'A1_L2_S', 'A1_L2_G', 'A1_L3', 'A1_L3_E_Imp', 'A1_L3_S_Imp', 'A1_L3_G_Imp', 'A1_L4', 'A1_L4_E', 'A1_L4_E_Red', 'A1_L4_E_Yellow', 'A1_L4_E_Green', 'A1_L4_S', 'A1_L4_S_Red', 'A1_L4_S_Yellow', 'A1_L4_S_Green', 'A1_L4_G', 'A1_L4_G_Red', 'A1_L4_G_Yellow', 'A1_L4_G_Green', 'a2_x1', 'a2_x2', 'A2_PRICE', 'a2_x3', 'A2_MAT_CC', 'A2_MAT_OC', 'A2_MAT_WO', 'A2_MAT_RM', 'A2_MAT_PE', 'a2_x4', 'A2_COM_CHN', 'A2_COM_EUR', 'A2_COM_BGD', 'A2_COM_VNM', 'A2_COM_A_G', 'a2_x5', 'A2_Q_LOW', 'A2_Q_MED', 'A2_Q_HIGH', 'a2_x6', 'A2_Label', 'A2_L1', 'A2_L1_E_CO2', 'A2_L1_E_WAT', 'A2_L1_E_WAS', 'A2_L1_E_ENE', 'A2_L1_S_WFS', 'A2_L1_S_HRS', 'A2_L1_G_MGS', 'A2_L1_G_CSR', 'A2_L2', 'A2_L2_E', 'A2_L2_S', 'A2_L2_G', 'A2_L3', 'A2_L3_E_Imp', 'A2_L3_S_Imp', 'A2_L3_G_Imp', 'A2_L4', 'A2_L4_E', 'A2_L4_E_Red', 'A2_L4_E_Yellow', 'A2_L4_E_Green', 'A2_L4_S', 'A2_L4_S_Red', 'A2_L4_S_Yellow', 'A2_L4_S_Green', 'A2_L4_G', 'A2_L4_G_Red', 'A2_L4_G_Yellow', 'A2_L4_G_Green', 'a3_x1', 'a3_x2', 'a3_x3', 'a3_x4', 'a3_x5', 'a3_x6']]

#####################################################################################################
# Prepare database for Biogeme
#####################################################################################################

            # Initialize database
            database = db.Database("Data", df)

            # Data organized as panel data. The variable Resp identifies each individual.
            database.panel("RID_New")

             # Define the name of the variables as Python variables: This command reads the first row of data as variable names.
            globals().update(database.variables)
        
        
#####################################################################################################
# Biogeme Parameter Definition
#####################################################################################################        



### Fixed parameters

            A1_ASC = Beta('A1_ASC', value[x][y][z][0], None, None, 0)
            A3_ASC = Beta('A3_ASC', value[x][y][z][1], None, None, 0)

        
### Random parameters

            # Mean estimates
            B_PRICE_mean = Beta('B_PRICE_mean', value[x][y][z][2], None, None, 0)
            B_MAT_OC_mean = Beta('B_MAT_OC_mean', value[x][y][z][3], None, None, 0)
            B_MAT_WO_mean = Beta('B_MAT_WO_mean', value[x][y][z][4], None, None, 0)
            B_MAT_RM_mean = Beta('B_MAT_RM_mean', value[x][y][z][5], None, None, 0)
            B_MAT_PE_mean = Beta('B_MAT_PE_mean', value[x][y][z][6], None, None, 0)
            B_COM_EUR_mean = Beta('B_COM_EUR_mean', value[x][y][z][7], None, None, 0)
            B_COM_BGD_mean = Beta('B_COM_BGD_mean', value[x][y][z][8], None, None, 0)
            B_COM_VNM_mean = Beta('B_COM_VNM_mean', value[x][y][z][9], None, None, 0)
            B_COM_A_G_mean = Beta('B_COM_A_G_mean', value[x][y][z][10], None, None, 0)
            B_Q_MED_mean = Beta('B_Q_MED_mean', value[x][y][z][11], None, None, 0)
            B_Q_HIGH_mean = Beta('B_Q_HIGH_mean', value[x][y][z][12], None, None, 0)
            
            if z == "WL":
                B_L1_E_CO2_mean = Beta('B_L1_E_CO2_mean', value[x][y][z][13], None, None, 0)
                B_L1_E_WAT_mean = Beta('B_L1_E_WAT_mean', value[x][y][z][14], None, None, 0)
                B_L1_E_WAS_mean = Beta('B_L1_E_WAS_mean', value[x][y][z][15], None, None, 0)
                B_L1_E_ENE_mean = Beta('B_L1_E_ENE_mean', value[x][y][z][16], None, None, 0)
                
                B_L1_S_WFS_mean = Beta('B_L1_S_WFS_mean', value[x][y][z][17], None, None, 0)
                B_L1_S_HRS_mean = Beta('B_L1_S_HRS_mean', value[x][y][z][18], None, None, 0)
                
                B_L1_G_MGS_mean = Beta('B_L1_G_MGS_mean', value[x][y][z][19], None, None, 0)
                B_L1_G_CSR_mean = Beta('B_L1_G_CSR_mean', value[x][y][z][20], None, None, 0)
                
                
                B_L2_mean = Beta('B_L2_mean', value[x][y][z][21], None, None, 0)
                
                B_L2_E_mean = Beta('B_L2_E_mean', value[x][y][z][22], None, None, 0)
                B_L2_S_mean = Beta('B_L2_S_mean', value[x][y][z][23], None, None, 0)
                B_L2_G_mean = Beta('B_L2_G_mean', value[x][y][z][24], None, None, 0)

                
                B_L3_mean = Beta('B_L3_mean', value[x][y][z][25], None, None, 0)
                
                B_L3_E_Imp_mean = Beta('B_L3_E_Imp_mean', value[x][y][z][26], None, None, 0)
                B_L3_S_Imp_mean = Beta('B_L3_S_Imp_mean', value[x][y][z][27], None, None, 0)
                B_L3_G_Imp_mean = Beta('B_L3_G_Imp_mean', value[x][y][z][28], None, None, 0)

                
                B_L4_mean = Beta('B_L4_mean', value[x][y][z][29], None, None, 0)
                
                B_L4_E_Yellow_mean = Beta('B_L4_E_Yellow_mean', value[x][y][z][30], None, None, 0)
                B_L4_E_Green_mean = Beta('B_L4_E_Green_mean', value[x][y][z][31], None, None, 0)

                B_L4_S_Yellow_mean = Beta('B_L4_S_Yellow_mean', value[x][y][z][32], None, None, 0)
                B_L4_S_Green_mean = Beta('B_L4_S_Green_mean', value[x][y][z][33], None, None, 0)

                B_L4_G_Yellow_mean = Beta('B_L4_G_Yellow_mean', value[x][y][z][34], None, None, 0)
                B_L4_G_Green_mean = Beta('B_L4_G_Green_mean', value[x][y][z][35], None, None, 0)
                
            else:
                None
    

            #Standard deviations estimates
            B_PRICE_STD = Beta('B_PRICE_STD', 0.0001, None, None, 0)
            B_MAT_OC_STD = Beta('B_MAT_OC_STD', 0.0001, None, None, 0)
            B_MAT_WO_STD = Beta('B_MAT_WO_STD', 0.0001, None, None, 0)
            B_MAT_RM_STD = Beta('B_MAT_RM_STD', 0.0001, None, None, 0)
            B_MAT_PE_STD = Beta('B_MAT_PE_STD', 0.0001, None, None, 0)
            B_COM_EUR_STD = Beta('B_COM_EUR_STD', 0.0001, None, None, 0)
            B_COM_BGD_STD = Beta('B_COM_BGD_STD', 0.0001, None, None, 0)
            B_COM_VNM_STD = Beta('B_COM_VNM_STD', 0.0001, None, None, 0)
            B_COM_A_G_STD = Beta('B_COM_A_G_STD', 0.0001, None, None, 0)
            B_Q_MED_STD = Beta('B_Q_MED_STD', 0.0001, None, None, 0)
            B_Q_HIGH_STD = Beta('B_Q_HIGH_STD', 0.0001, None, None, 0)
            
            if z == "WL":
                B_L1_E_CO2_STD = Beta('B_L1_E_CO2_STD', 0.0001, None, None, 0)
                B_L1_E_WAT_STD = Beta('B_L1_E_WAT_STD', 0.0001, None, None, 0)
                B_L1_E_WAS_STD = Beta('B_L1_E_WAS_STD', 0.0001, None, None, 0)
                B_L1_E_ENE_STD = Beta('B_L1_E_ENE_STD', 0.0001, None, None, 0)
                
                B_L1_S_WFS_STD = Beta('B_L1_S_WFS_STD', 0.0001, None, None, 0)
                B_L1_S_HRS_STD = Beta('B_L1_S_HRS_STD', 0.0001, None, None, 0)
                
                B_L1_G_MGS_STD = Beta('B_L1_G_MGS_STD', 0.0001, None, None, 0)
                B_L1_G_CSR_STD = Beta('B_L1_G_CSR_STD', 0.0001, None, None, 0)
                
                
                B_L2_STD = Beta('B_L2_STD', 0.0001, None, None, 0)
                
                B_L2_E_STD = Beta('B_L2_E_STD', 0.0001, None, None, 0)
                B_L2_S_STD = Beta('B_L2_S_STD', 0.0001, None, None, 0)
                B_L2_G_STD = Beta('B_L2_G_STD', 0.0001, None, None, 0)

                
                B_L3_STD = Beta('B_L3_STD', 0.0001, None, None, 0)
                
                B_L3_E_Imp_STD = Beta('B_L3_E_Imp_STD', 0.0001, None, None, 0)
                B_L3_S_Imp_STD = Beta('B_L3_S_Imp_STD', 0.0001, None, None, 0)
                B_L3_G_Imp_STD = Beta('B_L3_G_Imp_STD', 0.0001, None, None, 0)

                
                B_L4_STD = Beta('B_L4_STD', 0.0001, None, None, 0)
                
                B_L4_E_Yellow_STD = Beta('B_L4_E_Yellow_STD', 0.0001, None, None, 0)
                B_L4_E_Green_STD = Beta('B_L4_E_Green_STD', 0.0001, None, None, 0)

                B_L4_S_Yellow_STD = Beta('B_L4_S_Yellow_STD', 0.0001, None, None, 0)
                B_L4_S_Green_STD = Beta('B_L4_S_Green_STD', 0.0001, None, None, 0)

                B_L4_G_Yellow_STD = Beta('B_L4_G_Yellow_STD', 0.0001, None, None, 0)
                B_L4_G_Green_STD = Beta('B_L4_G_Green_STD', 0.0001, None, None, 0)

            else:
                None


            #Initialization 
            B_PRICE  = B_PRICE_mean  + B_PRICE_STD  * Dr1
            B_MAT_OC  = B_MAT_OC_mean  + B_MAT_OC_STD  * Dr2
            B_MAT_WO  = B_MAT_WO_mean  + B_MAT_WO_STD * Dr3
            B_MAT_RM  = B_MAT_RM_mean  + B_MAT_RM_STD * Dr4
            B_MAT_PE  = B_MAT_PE_mean  + B_MAT_PE_STD * Dr5
            B_COM_EUR = B_COM_EUR_mean + B_COM_EUR_STD * Dr6
            B_COM_BGD  = B_COM_BGD_mean  + B_COM_BGD_STD * Dr7
            B_COM_VNM  = B_COM_VNM_mean  + B_COM_VNM_STD * Dr8
            B_COM_A_G  = B_COM_A_G_mean  + B_COM_A_G_STD  * Dr9
            B_Q_MED  = B_Q_MED_mean  + B_Q_MED_STD  * Dr10
            B_Q_HIGH = B_Q_HIGH_mean + B_Q_HIGH_STD * Dr11
            
            if z == "WL":
                B_L1_E_CO2 = B_L1_E_CO2_mean + B_L1_E_CO2_STD * Dr12
                B_L1_E_WAT = B_L1_E_WAT_mean + B_L1_E_WAT_STD * Dr13
                B_L1_E_WAS = B_L1_E_WAS_mean + B_L1_E_WAS_STD * Dr14
                B_L1_E_ENE = B_L1_E_ENE_mean + B_L1_E_ENE_STD * Dr15
                
                B_L1_S_WFS = B_L1_S_WFS_mean + B_L1_S_WFS_STD * Dr16
                B_L1_S_HRS = B_L1_S_HRS_mean + B_L1_S_HRS_STD * Dr17
                
                B_L1_G_MGS = B_L1_G_MGS_mean + B_L1_G_MGS_STD * Dr18
                B_L1_G_CSR = B_L1_G_CSR_mean + B_L1_G_CSR_STD * Dr19
                
                
                B_L2 = B_L2_mean + B_L2_STD * Dr20
                
                B_L2_E = B_L2_E_mean + B_L2_E_STD * Dr21
                B_L2_S = B_L2_S_mean + B_L2_S_STD * Dr22
                B_L2_G = B_L2_G_mean + B_L2_G_STD * Dr23
                
                
                B_L3 = B_L3_mean + B_L3_STD * Dr24
                
                B_L3_E_Imp = B_L3_E_Imp_mean + B_L3_E_Imp_STD * Dr25
                B_L3_S_Imp = B_L3_S_Imp_mean + B_L3_S_Imp_STD * Dr26
                B_L3_G_Imp = B_L3_G_Imp_mean + B_L3_G_Imp_STD * Dr27
                
                
                B_L4 = B_L4_mean + B_L4_STD * Dr28
                
                B_L4_E_Yellow = B_L4_E_Yellow_mean + B_L4_E_Yellow_STD * Dr29
                B_L4_E_Green = B_L4_E_Green_mean + B_L4_E_Green_STD * Dr30
                
                B_L4_S_Yellow = B_L4_S_Yellow_mean + B_L4_S_Yellow_STD * Dr31
                B_L4_S_Green = B_L4_S_Green_mean + B_L4_S_Green_STD * Dr32
                
                B_L4_G_Yellow = B_L4_G_Yellow_mean + B_L4_G_Yellow_STD * Dr33
                B_L4_G_Green = B_L4_G_Green_mean + B_L4_G_Green_STD * Dr34
                
            else:  
                None

        
    
#####################################################################################################
# Utility function
#####################################################################################################   

            # Utility functions
            ## Without label
            if z == "WoL":
                
                V1 = A1_ASC + B_PRICE * A1_PRICE + B_MAT_OC * A1_MAT_OC + B_MAT_WO * A1_MAT_WO + B_MAT_RM * A1_MAT_RM + B_MAT_PE * A1_MAT_PE + B_COM_EUR * A1_COM_EUR + B_COM_BGD * A1_COM_BGD + B_COM_VNM * A1_COM_VNM + B_COM_A_G * A1_COM_A_G + B_Q_MED * A1_Q_MED + B_Q_HIGH * A1_Q_HIGH
                V2 =          B_PRICE * A2_PRICE + B_MAT_OC * A2_MAT_OC + B_MAT_WO * A2_MAT_WO + B_MAT_RM * A2_MAT_RM + B_MAT_PE * A2_MAT_PE + B_COM_EUR * A2_COM_EUR + B_COM_BGD * A2_COM_BGD + B_COM_VNM * A2_COM_VNM + B_COM_A_G * A2_COM_A_G + B_Q_MED * A2_Q_MED + B_Q_HIGH * A2_Q_HIGH
                V3 = A3_ASC
            
            ## with label
            else:
                V1 = A1_ASC + B_PRICE * A1_PRICE + B_MAT_OC * A1_MAT_OC + B_MAT_WO * A1_MAT_WO + B_MAT_RM * A1_MAT_RM + B_MAT_PE * A1_MAT_PE + B_COM_EUR * A1_COM_EUR + B_COM_BGD * A1_COM_BGD + B_COM_VNM * A1_COM_VNM + B_COM_A_G * A1_COM_A_G + B_Q_MED * A1_Q_MED + B_Q_HIGH * A1_Q_HIGH + B_L1_E_CO2 * A1_L1_E_CO2 + B_L1_E_WAT * A1_L1_E_WAT + B_L1_E_WAS * A1_L1_E_WAS + B_L1_E_ENE * A1_L1_E_ENE + B_L1_S_WFS * A1_L1_S_WFS + B_L1_S_HRS * A1_L1_S_HRS + B_L1_G_MGS * A1_L1_G_MGS + B_L1_G_CSR * A1_L1_G_CSR + B_L2 * A1_L2 + B_L2_E * A1_L2_E + B_L2_S * A1_L2_S + B_L2_G * A1_L2_G + B_L3 * A1_L3 + B_L3_E_Imp * A1_L3_E_Imp + B_L3_S_Imp * A1_L3_S_Imp + B_L3_G_Imp * A1_L3_G_Imp + B_L4 * A1_L4 + B_L4_E_Yellow * A1_L4_E_Yellow + B_L4_E_Green * A1_L4_E_Green + B_L4_S_Yellow * A1_L4_S_Yellow + B_L4_S_Green * A1_L4_S_Green + B_L4_G_Yellow * A1_L4_G_Yellow + B_L4_G_Green * A1_L4_G_Green
                V2 =          B_PRICE * A2_PRICE + B_MAT_OC * A2_MAT_OC + B_MAT_WO * A2_MAT_WO + B_MAT_RM * A2_MAT_RM + B_MAT_PE * A2_MAT_PE + B_COM_EUR * A2_COM_EUR + B_COM_BGD * A2_COM_BGD + B_COM_VNM * A2_COM_VNM + B_COM_A_G * A2_COM_A_G + B_Q_MED * A2_Q_MED + B_Q_HIGH * A2_Q_HIGH + B_L1_E_CO2 * A2_L1_E_CO2 + B_L1_E_WAT * A2_L1_E_WAT + B_L1_E_WAS * A2_L1_E_WAS + B_L1_E_ENE * A2_L1_E_ENE + B_L1_S_WFS * A2_L1_S_WFS + B_L1_S_HRS * A2_L1_S_HRS + B_L1_G_MGS * A2_L1_G_MGS + B_L1_G_CSR * A2_L1_G_CSR + B_L2 * A2_L2 + B_L2_E * A2_L2_E + B_L2_S * A2_L2_S + B_L2_G * A2_L2_G + B_L3 * A2_L3 + B_L3_E_Imp * A2_L3_E_Imp + B_L3_S_Imp * A2_L3_S_Imp + B_L3_G_Imp * A2_L3_G_Imp + B_L4 * A2_L4 + B_L4_E_Yellow * A2_L4_E_Yellow + B_L4_E_Green * A2_L4_E_Green + B_L4_S_Yellow * A2_L4_S_Yellow + B_L4_S_Green * A2_L4_S_Green + B_L4_G_Yellow * A2_L4_G_Yellow + B_L4_G_Green * A2_L4_G_Green
                V3 = A3_ASC

            # Associate the utility with the numbering of the alternative
            V = {1: V1, 2: V2, 3: V3}

            # Availability condition
            av = {1: 1, 2: 1, 3: 1} 
        

            
#####################################################################################################
# Definition of the model
#####################################################################################################   
            # Create the Biogeme object
            prob = models.logit(V,av,BEST)
            condP = PanelLikelihoodTrajectory(prob)

            # Create the Biogeme object
            biogeme = bio.BIOGEME(database, log(MonteCarlo(condP)))

            #Output Model
            biogeme.modelName = "20"+time+'_MMNL_'+x+"_"+y+"_"+z
            
            # Estimation of Parameters
            results = biogeme.estimate()


#####################################################################################################
# Output
#####################################################################################################   
            
            #Read the results
            pandasResults = results.getEstimatedParameters(onlyRobust=False)
            
            # Order to Betas
            pandasResults = pandasResults.join(df_Para_Order)
            
            #General statistics 
            General_stats = results.getGeneralStatistics()
            GS = pd.DataFrame.from_dict(General_stats)
            GS = GS.T
            GS = GS.rename(columns={0:'Value'})
            GS.index.names = ['Parameter']
            GS = GS[['Value']]
            
            
#####################################################################################################
# Save data in Excel file
#####################################################################################################   



            with pd.ExcelWriter(path_results, mode='a', if_sheet_exists='replace') as writer:
                pandasResults.sort_values('Order').to_excel(writer, sheet_name=x+"_"+y+"_"+z, index=True)
            
            
            with pd.ExcelWriter(path_results, mode='a', if_sheet_exists='overlay') as writer:
                GS.to_excel(writer, sheet_name=x+"_"+y+"_"+z, index=True, startcol=11)

```
